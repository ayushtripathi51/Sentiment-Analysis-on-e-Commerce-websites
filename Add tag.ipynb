```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import re
```

```python
df_data = pd.read_csv('C:\\Users\\Ayush\\Desktop\\MINING2.csv',engine='python')
```

```python
df_data2=[]
x=17
for index,i in df_data.iterrows():
    #print(i)
    try:
        if((re.match("^(.)*(Amazon|amazon)+(.)*",i[2])) or (re.match("^(.)*(Amazon|amazon)+(.)*",i[3]))):
            df_data2.append({'id':i[1],'text':i[2],'source':i[3],'date':i[4],'tag':"Amazon"})
        elif((re.match("^(.)*(Myntra|myntra)+(.)*",i[2])) or (re.match("^(.)*(Myntra|myntra)+(.)*",i[3]))):
            df_data2.append({'id':i[1],'text':i[2],'source':i[3],'date':i[4],'tag':"Myntra"})
        elif((re.match("^(.)*(Flipkart|flipkart)+(.)*",i[2])) or (re.match("^(.)*(Flipkart|flipkart)+(.)*",i[3]))):
            df_data2.append({'id':i[1],'text':i[2],'source':i[3],'date':i[4],'tag':"Flipkart"})
        elif((re.match("^(.)*(Jabong|jabong)+(.)*",i[2])) or (re.match("^(.)*(Jabong|jabong)+(.)*",i[3]))):
            df_data2.append({'id':i[1],'text':i[2],'source':i[3],'date':i[4],'tag':"Jabong"})
        else:
            df_data2.append({'id':i[1],'text':i[2],'source':i[3],'date':i[4],'tag':"eBay"})
    except Exception as e:
        pass
df_data3=DataFrame(columns=('id', 'text','source','date','tag'))
df_data3=df_data3.append(df_data2)
df_data3
```

```python
df_data3.to_csv("C:\\Users\\Ayush\\Desktop\\MINING3.csv")
```
